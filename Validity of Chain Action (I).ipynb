```python
#Calculating old action(S) and new action (S') for each number of elements in the range 

%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import math
import random

A=int(input())
a4=[]
a41=[]
q=range(300,2200,200) #number of elements

for j in range(1,A+1):
    a2=[]
    a3=[]
    t1=-5
             
    for N in q:
        m=[]
        if N==q[0]:
            m.append(E[0])
        if N==q[1]:
            m.append(E[1])
        if N==q[2]:
            m.append(E[2])
        if N==q[3]:
            m.append(E[3])
        if N==q[4]:
            m.append(E[4])
        if N==q[5]:
            m.append(E[5])
        if N==q[6]:
            m.append(E[6])
        if N==q[7]:
            m.append(E[7])
        if N==q[8]:
            m.append(E[8])
        if N==q[9]:
            m.append(E[9])   
            
        def Tim(k):
            H=100
            rho=2000000
            a=1
            c=25
            b=10*(math.exp((k*H**2)/(2*rho))-1)
            d=b**2-4*a*c
            R= (-b+math.sqrt(d))/(2*a)
            return R
    

        T=Tim(N)
        x2=(t1-T)/2
        x1=-x2
        L=[]
        L0=[]
        count=0
        count1=0
        
        while count<N:
            R1 = np.random.uniform(0,1)
            R2 = np.random.uniform(0,1)
            R3= np.random.uniform(0,1)
            R4=np.random.uniform(0,1)
            lx=R1*(x2-x1)+x1
            lt=(T*t1)/(R2*(t1-T)+T)
            lx1=R1*(x2-x1)+x1
            lt1=(T*t1)/(R2*(t1-T)+T)
            if (np.square((lt)-(t1))>np.square(lx)and np.square((lt)-(T))>np.square(lx)):
                count+=1
                tuple=[lx,lt]
                L.append(tuple)
            if (np.square((lt1)-(t1))>np.square(lx1)and np.square((lt1)-(T))>np.square(lx1)):
                count1+=1
                tuple=[lx1,lt1]
                L0.append(tuple)
            
        x=[0,T]
        y=[0,t1]
        u=list(np.concatenate(m))
        
        L.append(x)
        L.append(y)
        
        L0.append(x)
        L0.append(y)
        #Add new element to L0 list
        L0.append(u)
        
        #Sort both L and L0 in the order of increasing time
        def takeSecond(elem):
            return elem[1]    
        L.sort(key=takeSecond) 
        
        def takeSecond(elem):
            return elem[1]    
        L0.sort(key=takeSecond) 
        
        #Old relation matrix
        R=[]
        for j in range(len(L)):
            R.append([])
            for i in range(len(L)):
                if (np.square(L[i][1]-L[j][1])>= np.square(L[i][0]-L[j][0]) and L[i][1]> L[j][1]):
                    R[j].append(1) 
                else:
                    R[j].append(0)
        I=np.identity(len(R))
        R1=np.dot(I,R)
        R2=np.dot(R,R)
        R3=np.dot(R,R2)
        c3=R3[0][len(R2)-1]
        Act=36*((4*c3/(N*(N-1))-1)) #Old Action (S)
        
        #New Relations matrix with an extra point
        Y4=[]
        for j in range(len(L0)):
            Y4.append([])
            for i in range(len(L0)):
                if (np.square(L0[i][1]-L0[j][1])>= np.square(L0[i][0]-L0[j][0]) and L0[i][1]> L0[j][1]):
                    Y4[j].append(1) 
                else:
                    Y4[j].append(0)
        I=np.identity(len(Y4))
        Y5=np.dot(I,Y4)
        Y6=np.dot(Y4,Y4)
        Y7=np.dot(Y4,Y6)
        c7=Y7[0][len(Y6)-1]
        Act1=36*((4*c7/(N*(N-1))-1)) #New Action (S')
        
        a3.append(Act1)
        a2.append(Act)
    a4.append(a2)
    a41.append(a3)

k1=np.sum(a4,0)/A  #Averaging action from 20 different causal sets 
k2=np.sum(a41,0)/A
w1=[a_i - b_i for a_i, b_i in zip(k2, k1)] #Taking the difference (delta S)
print(k1)
print(k2)
print(w1)
```
